planets.head(3)

```python
while json['next'] != None:
    
    print(json['next'])
    response = req.get(json['next'])
    json= response.json()
    planets = pd.concat(planets, pd.DataFrame(json['results'], axis=0, 
                                             ignore_index=True))    
```
